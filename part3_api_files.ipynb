```python
# ============================================================
# TASK 2 - API Integration
# ============================================================

import requests

# --- Step 1: Fetch 20 products and print table ---

print("\n--- 20 Products ---")

try:
    response = requests.get("https://dummyjson.com/products?limit=20", timeout=5)
    data = response.json()
    products = data["products"]

    print("ID   | Title                          | Category       | Price    | Rating")
    print("-" * 75)

    for product in products:
        print(str(product["id"]) + "    | " + product["title"][:30] + " | " + product["category"][:14] + " | $" + str(product["price"]) + " | " + str(product["rating"]))

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Something went wrong: " + str(e))


# --- Step 2: Filter rating >= 4.5 and sort by price descending ---

print("\n--- Products rated 4.5 or above (sorted by price high to low) ---")

try:
    response = requests.get("https://dummyjson.com/products?limit=20", timeout=5)
    data = response.json()
    products = data["products"]

    # Filter: only keep products with rating >= 4.5
    filtered = []
    for product in products:
        if product["rating"] >= 4.5:
            filtered.append(product)

    # Sort by price descending using bubble sort
    for i in range(len(filtered)):
        for j in range(i + 1, len(filtered)):
            if filtered[j]["price"] > filtered[i]["price"]:
                temp = filtered[i]
                filtered[i] = filtered[j]
                filtered[j] = temp

    for product in filtered:
        print(product["title"] + " — $" + str(product["price"]) + " — Rating: " + str(product["rating"]))

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Something went wrong: " + str(e))


# --- Step 3: Fetch laptops category ---

print("\n--- Laptops ---")

try:
    response = requests.get("https://dummyjson.com/products/category/laptops", timeout=5)
    data = response.json()
    laptops = data["products"]

    for laptop in laptops:
        print(laptop["title"] + " — $" + str(laptop["price"]))

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Something went wrong: " + str(e))


# --- Step 4: POST a new product ---

print("\n--- Adding new product ---")

new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

try:
    response = requests.post("https://dummyjson.com/products/add", json=new_product, timeout=5)
    print("Server response:")
    print(response.json())

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Something went wrong: " + str(e))
```

```python
# ============================================================
# TASK 3 - Exception Handling
# ============================================================

# --- Part A: Safe divide function ---

def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print("\n--- Safe Divide Tests ---")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


# --- Part B: Safe file reader function ---

def read_file_safe(filename):
    try:
        file = open(filename, "r", encoding="utf-8")
        content = file.read()
        file.close()
        return content
    except FileNotFoundError:
        print("Error: File '" + filename + "' not found.")
        return None
    finally:
        print("File operation attempt complete.")

print("\n--- Safe File Reader Tests ---")
print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))


# --- Part D: Input validation loop with API lookup ---

print("\n--- Product Lookup ---")

while True:
    user_input = input("Enter a product ID (1-100) or 'quit' to exit: ")

    # Stop loop if user types quit
    if user_input.lower() == "quit":
        print("Goodbye.")
        break

    # Check it is a number
    if user_input.isnumeric() == False:
        print("Warning: Please enter a number.")
        continue

    # Convert to integer
    product_id = int(user_input)

    # Check it is in range
    if product_id < 1 or product_id > 100:
        print("Warning: Number must be between 1 and 100.")
        continue

    # Make the API call
    try:
        response = requests.get("https://dummyjson.com/products/" + str(product_id), timeout=5)

        if response.status_code == 404:
            print("Product not found.")

        elif response.status_code == 200:
            product = response.json()
            print("Title: " + product["title"] + "  Price: $" + str(product["price"]))

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")

    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")

    except Exception as e:
        print("Something went wrong: " + str(e))
```

```python
# ============================================================
# TASK 4 - Logging errors to file
# ============================================================

from datetime import datetime

# --- Write a log entry to error_log.txt ---

def log_error(location, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = "[" + timestamp + "] ERROR in " + location + ": " + error_type + " — " + message + "\n"

    log_file = open("error_log.txt", "a", encoding="utf-8")
    log_file.write(log_entry)
    log_file.close()

    print("Logged: " + log_entry.strip())


# --- Trigger 1: ConnectionError using unreachable URL ---

print("\n--- Triggering ConnectionError ---")

try:
    response = requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)

except requests.exceptions.ConnectionError:
    print("Connection failed.")
    log_error("fetch_products", "ConnectionError", "No connection could be made")

except requests.exceptions.Timeout:
    print("Request timed out.")
    log_error("fetch_products", "Timeout", "Request timed out")


# --- Trigger 2: HTTP 404 using a product ID that does not exist ---

print("\n--- Triggering 404 error ---")

try:
    response = requests.get("https://dummyjson.com/products/999", timeout=5)

    if response.status_code != 200:
        print("Product not found. Status: " + str(response.status_code))
        log_error("lookup_product", "HTTPError", "404 Not Found for product ID 999")

except requests.exceptions.ConnectionError:
    print("Connection failed.")
    log_error("lookup_product", "ConnectionError", "Could not connect")


# --- Print full contents of error_log.txt ---

print("\n--- Contents of error_log.txt ---")

log_file = open("error_log.txt", "r", encoding="utf-8")
print(log_file.read())
log_file.close()
```
